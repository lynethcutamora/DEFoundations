```python
import requests
from bs4 import BeautifulSoup

url='https://github.com/lynethcutamora/'
response=requests.get(url)

html_content=response.content
print(html_content)
```

```python
soup=BeautifulSoup(html_content,'html.parser')
h1_tags = soup.find('h1')
print(h1_tags.text)
```

```python
imgs = soup.find_all('img')

for img in imgs:
  print(img['src'])
```

```python
mylists = soup.find('ul')
print(mylists.text)
```

```python
text = mylists.text

with open('webscrapping_output.txt', 'w') as file:
  file.write(text)
```
