```python
import pandas as pd
import matplotlib.pyplot as plt

try:
  data={'Date': ['2023-08-01','2023-08-02','2023-08-03'],
        'Price': [100,105,110]}

  df=pd.DataFrame(data)
  df['Date']=pd.to_datetime(df['Date'])

  df.plot(x='Date',y='Price',kind='line')
  plt.xlabel('Date')
  plt.ylabel('Price')

  plt.show()

except Exception as e:
  print("error: ", e)
```

```python
try:
  data={'Product':['A','B','C'],
        'Sales': [500,750,300]}
  df=pd.DataFrame(data)

  df.plot(x="Product",y="Sales", kind="bar")
  plt.xlabel('Product')
  plt.ylabel('Sales')

  plt.show
except Exception as e:
  print("error: ",e)
```

```python
try:
    data={'Hours_Studied': [2,4,5,7,8,10],
          'Exam_Score': [60,70,75,85,90,95]}
    df=pd.DataFrame(data)

    df.plot(x="Hours_Studied",y="Exam_Score", kind="scatter")
    plt.xlabel('Hours_Studied')
    plt.ylabel('Exam_Score')

    plt.show
except Exception as e:
  print("error: ",e)
```

```python
data={'Toppings':['Pepperoni','Mushroom','Olive','Sausage'],
      'Count':[25,15,10,20]}
df=pd.DataFrame(data)

df.plot(x="Toppings", y="Count", kind="pie")
plt.show()
```
